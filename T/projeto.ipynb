# Objetivo 1 - Taxa de mortalidade
O primeiro objetivo é calcular a taxa de mortalidade durante o período de estudo proposto para observar se houve uma notável variação ao longo dos anos. Assim, podendo observar entre os municípios da baixada santista quais foram os mais afetados pelos acidentes.

```python
municipios = df_sinistros['MUNICIPIO'].unique()
tipo = df_sinistros['TIPO'].unique()
def select_municipio(municipio, tipo, ano):
    new_df_sinistros = df_sinistros[(df_sinistros['MUNICIPIO'] == municipio) & (df_sinistros['TIPO'] == tipo)]
    new_df_populacao = df_populacao[(df_populacao['Municipio'] == municipio) & (df_populacao['Ano'] == ano)] 
    return new_df_sinistros, new_df_populacao    

new_table = []
table_total = []
for ano in range(2001, 2021):
    coluna = 'MORTES_' + str(ano)
    for cidade in municipios:
        for tipos in tipo:
            resposta = select_municipio(cidade, tipos, ano)
            taxa_mortalidade = ((resposta[0].iloc[0][coluna] * 100000)/resposta[1].iloc[0]['PopulacaoResidente'])/1000
            taxa_mortalidade = round(taxa_mortalidade, 1)
            mortes = resposta[0].iloc[0][coluna]
            #print(f'{cidade} | {tipos} | {ano} | {taxa_mortalidade} | {mortes}')
            variaveis = [cidade, tipos, mortes, taxa_mortalidade, ano]
            if tipos != 'Total':
                new_table.append(variaveis)
            else:
                table_total.append(variaveis)
#print(new_table)
df_taxa_mortalidade = pd.DataFrame(new_table, columns=['Municipio', 'Tipo',  'Mortes', 'Taxa_mortalidade', 'Ano'])
df_taxa_mortalidade_total = pd.DataFrame(table_total, columns=['Municipio', 'Tipo',  'Mortes', 'Taxa_mortalidade', 'Ano'])
display(df_taxa_mortalidade)
display(df_taxa_mortalidade_total)
```
